```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./Resources/Photos/cat.jpg")
plt.imshow(img)
```

```python
normal = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(normal)
```

```python
def get_image(image, cmap=None):
    plt.axis("off")
    plt.imshow(image, cmap = cmap)
    plt.show()
```

```python
get_image(normal)
```

```python
#Colorized 
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
#gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
get_image(gray, cmap = "gray")
```

```python
#Blurring
blurred = cv2.blur(normal, ksize=(15,15))

get_image(blurred)
```

```python
#Binary 
v, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
get_image(binary, cmap="gray")
```

```python
#invert 
invert = cv2.bitwise_not(binary)
get_image(invert, cmap="gray")
```
